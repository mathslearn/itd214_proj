```python
# Do bar graph for each year where vertical axis is sum of number of rows for df["reviews.rating'] and horizontal axis is year.
# Colour within each bar the different df['reviews.rating'].
# Label sum total for each df['reviews.rating'] in the middle of coloured section.
# Label sum total for total number of rows for df["reviews.rating'] at top of bar.

import matplotlib.pyplot as plt

# Group data by year and rating, then count the occurrences
rating_counts = df.groupby(['year', 'reviews.rating'])['reviews.rating'].count().unstack()

# Create the bar plot
rating_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews per Rating per Year')
plt.xticks(rotation=0)
plt.legend(title='Rating')

# Annotate the bars with the total count for each year
for i, year in enumerate(rating_counts.index):
    total_count = rating_counts.loc[year].sum()
    plt.text(i, total_count, str(int(total_count)), ha='center', va='bottom')

# Annotate each segment with its count, handling NaN values
for i, year in enumerate(rating_counts.index):
    bottom = 0
    for rating in rating_counts.columns:
        count = rating_counts.loc[year, rating]
        # Check if count is NaN and replace with 0 for annotation
        if pd.isna(count):
            count = 0
        plt.text(i, bottom + count / 2, str(int(count)), ha='center', va='center')
        bottom += count

plt.tight_layout()
plt.show()
```

```python
# Do bar graph for each year where vertical axis is percentage of number of rows for df["reviews.rating'] and horizontal axis is year.
# Colour within each bar the different df['reviews.rating'].
# Label percentage total for each df['reviews.rating'] in the middle of coloured section.
# Label percentage total for total number of rows for df["reviews.rating'] and sum total for total number of rows for df["reviews.rating'] at top of bar.

import matplotlib.pyplot as plt

# Group data by year and rating, then calculate the percentage of each rating for each year
rating_percentages = df.groupby(['year', 'reviews.rating'])['reviews.rating'].count().unstack()
rating_percentages = rating_percentages.div(rating_percentages.sum(axis=1), axis=0) * 100

# Create the bar plot
ax = rating_percentages.plot(kind='bar', stacked=True, figsize=(10, 6))

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Percentage of Reviews')
plt.title('Percentage of Reviews per Rating per Year')
plt.xticks(rotation=0)
plt.legend(title='Rating')

# Annotate each segment with its percentage
for i, year in enumerate(rating_percentages.index):
    bottom = 0
    for rating in rating_percentages.columns:
        percentage = rating_percentages.loc[year, rating]
        if pd.notna(percentage):  # Check for NaN values
            plt.text(i, bottom + percentage / 2, f'{percentage:.1f}%', ha='center', va='center')
        bottom += percentage

# Annotate the bars with the total percentage for each year (which should always be 100%)
for i, year in enumerate(rating_percentages.index):
    total_percentage = rating_percentages.loc[year].sum()
    plt.text(i, total_percentage, f'{total_percentage:.1f}%', ha='center', va='bottom')

# Calculate and display the total number of reviews for each year at the top of the bar
total_reviews_per_year = df.groupby('year')['reviews.rating'].count()
for i, year in enumerate(rating_percentages.index):
    total_count = total_reviews_per_year[year]
    ax.text(i, rating_percentages.loc[year].sum() + 2, str(int(total_count)), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# No undersampling.

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score
import pandas as pd
import numpy as np

# Initialize lists to store results
results = []  # Store results for each model as dictionaries

# Define skf before using it
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Loop through the models
for name, model in models:
    print(f"Evaluating model: {name}")

    # Initialize lists to store metrics for cross-validation
    model_cv_scores = []

    # Cross-validation loop to calculate CV accuracy scores
    for train_index, test_index in skf.split(x, y):
        # Split data into training and testing sets
        x_train, x_test = x.iloc[train_index], x.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Fit model on the original training data (no undersampling)
        model.fit(x_train, y_train)

        # Make predictions for the test set of the fold
        y_test_pred = model.predict(x_test)

        # Calculate and store test accuracy (CV)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        model_cv_scores.append(test_accuracy)

    # After cross-validation, calculate the mean CV accuracy
    mean_cv_accuracy = np.mean(model_cv_scores)

    # Train the model on the entire training data
    model.fit(x_train, y_train)

    # Calculate metrics for the entire training data (outside of the CV loop)
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)

    # Calculate train accuracy, test accuracy, train precision, and test precision
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    train_precision = precision_score(y_train, y_train_pred, average='weighted')
    test_precision = precision_score(y_test, y_test_pred, average='weighted')

    # Store results for this model as a dictionary
    results.append({
        'Model': name,
        'Cross-Validation Accuracy Scores': [round(score, 4) for score in model_cv_scores],  # Store CV scores from folds
        'Mean CV Accuracy': round(mean_cv_accuracy, 4),  # Store mean CV accuracy
        'Train Accuracy': round(train_accuracy, 4),  # Train accuracy (after entire training data)
        'Test Accuracy': round(test_accuracy, 4),  # Test accuracy (on hold-out test set)
        'Train Precision': round(train_precision, 4),  # Train precision (after entire training data)
        'Test Precision': round(test_precision, 4)   # Test precision (on hold-out test set)
    })

# Create a DataFrame from the results
df_results = pd.DataFrame(results)

# Display the results
print(df_results)
```

```python
# Undersampling of the majority class (sentiment=1) in the training set.

from sklearn.model_selection import StratifiedKFold
from sklearn.utils import resample  # Import the resample function from sklearn.utils
from sklearn.metrics import accuracy_score, precision_score
import pandas as pd
import numpy as np

# Initialize lists to store results
results = []  # Store results for each model as dictionaries

# Define skf before using it
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Loop through the models
for name, model in models:
    print(f"Evaluating model: {name}")

    # Initialize lists to store metrics for cross-validation
    model_cv_scores = []

    # Cross-validation loop to calculate CV accuracy scores
    for train_index, test_index in skf.split(x, y):
        # Split data into training and testing sets
        x_train, x_test = x.iloc[train_index], x.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Undersample the majority class (sentiment=1) in the training set
        x_train_resampled, y_train_resampled = resample(
            x_train[y_train == 1], y_train[y_train == 1],
            replace=False, n_samples=(y_train == 0).sum(), random_state=42
        )
        x_train_resampled = pd.concat([x_train[y_train == 0], x_train_resampled])
        y_train_resampled = pd.concat([y_train[y_train == 0], y_train_resampled])

        # Fit model on the resampled data
        model.fit(x_train_resampled, y_train_resampled)

        # Make predictions for the test set of the fold
        y_test_pred = model.predict(x_test)

        # Calculate and store test accuracy (CV)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        model_cv_scores.append(test_accuracy)

    # After cross-validation, calculate the mean CV accuracy
    mean_cv_accuracy = np.mean(model_cv_scores)

    # Train the model on the entire resampled training data
    model.fit(x_train_resampled, y_train_resampled)

    # Calculate metrics for the entire training data (outside of the CV loop)
    y_train_pred = model.predict(x_train_resampled)
    y_test_pred = model.predict(x_test)

    # Calculate train accuracy, test accuracy, train precision, and test precision
    train_accuracy = accuracy_score(y_train_resampled, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    train_precision = precision_score(y_train_resampled, y_train_pred, average='weighted')
    test_precision = precision_score(y_test, y_test_pred, average='weighted')

    # Store results for this model as a dictionary
    results.append({
        'Model': name,
        'Cross-Validation Accuracy Scores': [round(score, 4) for score in model_cv_scores],  # Store CV scores from folds
        'Mean CV Accuracy': round(mean_cv_accuracy, 4),  # Store mean CV accuracy
        'Train Accuracy': round(train_accuracy, 4),  # Train accuracy (after entire training data)
        'Test Accuracy': round(test_accuracy, 4),  # Test accuracy (on hold-out test set)
        'Train Precision': round(train_precision, 4),  # Train precision (after entire training data)
        'Test Precision': round(test_precision, 4)   # Test precision (on hold-out test set)
    })

# Create a DataFrame from the results
df_results = pd.DataFrame(results)

# Display the results
print(df_results)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.arima.model import ARIMA

# Safe MAPE calculation to avoid division by zero or very small values
def safe_mape(y_true, y_pred, epsilon=1e-10):
    # Avoid division by zero by adding a small epsilon
    return np.mean(np.abs((y_true - y_pred) / (y_true + epsilon))) * 100

# Function to evaluate forecast performance
def evaluate_forecast(y_true, y_pred):
    # MAE (Mean Absolute Error)
    mae = mean_absolute_error(y_true, y_pred)
    # MSE (Mean Squared Error)
    mse = mean_squared_error(y_true, y_pred)
    # RMSE (Root Mean Squared Error)
    rmse = np.sqrt(mse)
    # Safe MAPE (Mean Absolute Percentage Error)
    mape = safe_mape(y_true, y_pred)

    return mae, mse, rmse, mape

# Dictionary to store results for each model
model_results = {
    'daily': {
        'model': None,
        'mae': None,
        'mse': None,
        'rmse': None,
        'mape': None
    },
    'weekly': {
        'model': None,
        'mae': None,
        'mse': None,
        'rmse': None,
        'mape': None
    },
    'monthly': {
        'model': None,
        'mae': None,
        'mse': None,
        'rmse': None,
        'mape': None
    }
}

# 1. **Daily Model** - Fit the model and store the results
# Assuming 'df3' is your daily DataFrame and 'y_true_daily' is the true sentiment values for the test set

y_daily = df3['sentiment']
model_daily = ARIMA(y_daily, order=(1, 1, 1))
fitted_model_daily = model_daily.fit()
model_results['daily']['model'] = fitted_model_daily

# Get predicted values from the fitted model (adjust as needed)
start_date_daily = df3.index[0]
end_date_daily = df3.index[-1]
y_pred_daily = fitted_model_daily.predict(start=start_date_daily, end=end_date_daily, dynamic=False)

# Assuming y_true_daily is the actual values for the test set
y_true_daily = df3['sentiment'].dropna()  # Drop NaN values for consistency
y_pred_daily = y_pred_daily[:len(y_true_daily)]  # Ensure same length for comparison

# Evaluate the daily forecast
mae_daily, mse_daily, rmse_daily, mape_daily = evaluate_forecast(y_true_daily, y_pred_daily)
model_results['daily']['mae'] = mae_daily
model_results['daily']['mse'] = mse_daily
model_results['daily']['rmse'] = rmse_daily
model_results['daily']['mape'] = mape_daily

# 2. **Weekly Model** - Fit the model and store the results
# Assuming 'df4' is your weekly DataFrame and 'y_true_weekly' is the true sentiment values for the test set

y_weekly = df4['sentiment']
model_weekly = ARIMA(y_weekly, order=(1, 1, 1))
fitted_model_weekly = model_weekly.fit()
model_results['weekly']['model'] = fitted_model_weekly

# Get predicted values for weekly model (adjust as necessary)
start_date_weekly = df4.index[0]
end_date_weekly = df4.index[-1]
y_pred_weekly = fitted_model_weekly.predict(start=start_date_weekly, end=end_date_weekly, dynamic=False)

# Assuming y_true_weekly is the actual values for the test set
y_true_weekly = df4['sentiment'].dropna()  # Drop NaN values for consistency
y_pred_weekly = y_pred_weekly[:len(y_true_weekly)]  # Ensure same length for comparison

# Evaluate the weekly forecast
mae_weekly, mse_weekly, rmse_weekly, mape_weekly = evaluate_forecast(y_true_weekly, y_pred_weekly)
model_results['weekly']['mae'] = mae_weekly
model_results['weekly']['mse'] = mse_weekly
model_results['weekly']['rmse'] = rmse_weekly
model_results['weekly']['mape'] = mape_weekly

# 3. **Monthly Model** - Fit the model and store the results
# Assuming 'df5' is your monthly DataFrame and 'y_true_monthly' is the true sentiment values for the test set

y_monthly = df5['sentiment']
model_monthly = ARIMA(y_monthly, order=(1, 1, 1))
fitted_model_monthly = model_monthly.fit()
model_results['monthly']['model'] = fitted_model_monthly

# Get predicted values for monthly model (adjust as necessary)
start_date_monthly = df5.index[0]
end_date_monthly = df5.index[-1]
y_pred_monthly = fitted_model_monthly.predict(start=start_date_monthly, end=end_date_monthly, dynamic=False)

# Assuming y_true_monthly is the actual values for the test set
y_true_monthly = df5['sentiment'].dropna()  # Drop NaN values for consistency
y_pred_monthly = y_pred_monthly[:len(y_true_monthly)]  # Ensure same length for comparison

# Evaluate the monthly forecast
mae_monthly, mse_monthly, rmse_monthly, mape_monthly = evaluate_forecast(y_true_monthly, y_pred_monthly)
model_results['monthly']['mae'] = mae_monthly
model_results['monthly']['mse'] = mse_monthly
model_results['monthly']['rmse'] = rmse_monthly
model_results['monthly']['mape'] = mape_monthly

# Print the results for all models
print("Model Performance Metrics:")
for model, metrics in model_results.items():
    print(f"\n{model.capitalize()} Model:")
    print(f"  MAE: {metrics['mae']}")
    print(f"  MSE: {metrics['mse']}")
    print(f"  RMSE: {metrics['rmse']}")
    print(f"  MAPE: {metrics['mape']}")

# Optionally, you can store the results in a DataFrame for easier comparison
df_results = pd.DataFrame.from_dict(model_results, orient='index')
print(df_results)
```
